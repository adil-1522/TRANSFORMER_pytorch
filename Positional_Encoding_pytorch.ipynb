**Class**
Lets combine all the code into a cute class

```python
import torch
import torch.nn as nn


class PositionalEncoding(nn.Module):

  def __init__(self,d_model,sequence_length):
    super().__init__()
    self.d_model = d_model
    self.sequence_length = sequence_length

  def forward(self):
    even_i = torch.arange(0,self.d_model,2)
    denominator = torch.pow(10000,even_i/d_model)
    position = torch.arange(self.sequence_length).reshape(self.sequence_length,1)
    even_PE = torch.sin(position/denominator)
    odd_PE = torch.cos(position/denominator)
    stacked = torch.stack([even_PE,odd_PE],dim=2)
    PE = torch.flatten(stacked,start_dim=1,end_dim=2)
    return PE
```

```python
pe = PositionalEncoding(d_model=6, sequence_length=10)
pe.forward()
```
